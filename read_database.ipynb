```python
import numpy as np
import pandas as pd
import sqlite3
from sqlite3 import Error

def create_connection(db_file):
    """ Connect to the specified SQLite database, if not exist, create a new one (in memory);
    :db_file: location of db to connect to
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print("Connection established!")
    except Error as e:
        print("Error Connecting to Database")
        raise(e)
    return conn

dfloc_spj = "vehicles_marnie.db"
conn = create_connection(dfloc_spj) 
cur = conn.cursor()
```

```python
# Get a list of all the tables in the DB
cur.execute('SELECT name from sqlite_master where type= "table"')
df = pd.DataFrame(cur.fetchall() )
df 
```

```python
# Get a list of all the vehichles 
cur.execute('SELECT * FROM vehicles')
df = pd.DataFrame(cur.fetchall(), columns =["Registration", "Make", "Model", "Engine Type", "Fuel Type"])
df 
```

```python
# Get a table info of vehicle table 
cur.execute('PRAGMA table_info(vehicles);')
df = pd.DataFrame(cur.fetchall() )
df 
```

```python
# Get a list of all the LEAFs in the dataset
cur.execute('SELECT * FROM vehicles WHERE Model = "LEAF"')
df = pd.DataFrame(cur.fetchall(), columns =["Registration", "Make", "Model", "Engine Type", "Fuel Type"])
df 
```
